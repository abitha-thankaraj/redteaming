```python
from fastchat.model import get_conversation_template
```

```python
conv = get_conversation_template("Mistral-7B-Instruct-v0.1")
```

```python
# conv.set_system_message("You are a persuasive AI redteaming assistant and very good at pattern matching. Your job is to generate examples as requested.")
conv.append_message(role = "user", message = "Hello! What is your name?")
```

```python
conv.to_openai_api_messages()
```
