```python
import os, glob
from redteam.utils.data_utils import *
```

```python
data_dir = "/data/group_data/rl/datasets/redteaming/gen_multiturn_prompts/"
input_fnames = glob.glob(f"{data_dir}*.json")
```

```python
chunk_size = 100

for fname in input_fnames:
    data = read_json(fname)
    f = fname.split('/')[-1]
    dataset_name = f.split('_')[1] + "_chunked"
    print(dataset_name)
    os.makedirs(os.path.join(data_dir, dataset_name))
    for i in range(0, len(data), chunk_size):
        start_idx = i
        end_idx = min(i+chunk_size, len(data))
        ds = data[start_idx : end_idx]
        write_json(ds, os.path.join(data_dir, dataset_name, f.replace(".json", f"_{start_idx}_{end_idx}.json")))
        
        print(i, )
        
    
    print(f, len(data))
```
