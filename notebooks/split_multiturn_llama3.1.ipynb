```python
# Initialize the LLaMA tokenizer
tokenizer = tokenizer = transformers.AutoTokenizer.from_pretrained(model_name, 
                                                       config=config,
                                                       trust_remote_code=True,
                                                       model_max_length=64,
                                                       padding_side="right",
                                                       use_fast=False)

# Define a sample conversation
conversation = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Hello, how are you?"},
    {"role": "assistant", "content": "I'm doing well, thank you! How can I assist you today?"},
    {"role": "user", "content": "What's the weather like today?"},
    {"role": "assistant", "content": "I'm doing well, thank you! How can I assist you today?"},

]

def mask_non_user_tokens(tokens, conversation):
    masked_tokens = [tokenizer.pad_token_id] * len(tokens)  # Start with all tokens masked
    user_prefix = "<|start_header_id|>user<|end_header_id|>\n\n"
    user_suffix = "<|eot_id|>"
    user_prefix_ids = tokenizer.encode(user_prefix, add_special_tokens=False)
    user_contents = [user_prefix + msg['content']+ user_suffix for msg in conversation if msg['role'] == 'user']

    start_index = 0
    for content in user_contents:
        content_tokens = tokenizer.encode(content, add_special_tokens=False)
        print(content_tokens)
        print(tokenizer.decode(content_tokens))
        # Find user prefix
        while True:
            try:
                start_index = tokens.index(content_tokens[0], start_index)
                if tokens[start_index:start_index+len(content_tokens)] == content_tokens:
                    masked_tokens[start_index+len(user_prefix_ids):start_index+len(content_tokens)] = tokens[start_index+len(user_prefix_ids):start_index+len(content_tokens)]
                    break
                start_index += 1
            except ValueError:
                break    
    return masked_tokens


# finetune_right_pad_id

tokenizer.pad_token_id = 128004
tokenizer.mask_token_id = 128004
tokens = tokenizer.apply_chat_template(conversation, tokenize=True, padding="max_length", )

# Mask non-user tokens
masked_tokens = mask_non_user_tokens(tokens, conversation)

# Decode masked tokens
masked_text = tokenizer.decode(masked_tokens)

print("Masked text (only user messages visible):")
print(masked_text)
print("\n")
print(masked_text.replace("<|finetune_right_pad_id|>", ""))
```
